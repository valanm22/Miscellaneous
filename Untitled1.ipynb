```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset
import torchvision
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import math
from PIL import Image
```

```python
NUM_EPOCHS = 20
LEANRING_RATE = 0.001
BATCH_SIZE = 64
```

```python
# transform = transforms.ToTensor()
transform = transforms.Compose([transforms.Resize((256, 256)),
                                transforms.ToTensor(),
                                transforms.Normalize((0.485, 0.456, 0.406),
                                                     (0.229, 0.224, 0.225))])

target_transform = transforms.Compose([transforms.Resize((256, 256)),
                                transforms.PILToTensor()])

train = datasets.OxfordIIITPet(root='./data', split='trainval',
                          target_types='segmentation',
                          transform=transform,
                          target_transform=target_transform,
                          download=True)

test = datasets.OxfordIIITPet(root='./data', split='test',
                              transform=transform,
                              target_transform=target_transform,
                              target_types='segmentation',
                              download=True)
```

```python
#Creating dataloaders
train_loader = torch.utils.data.DataLoader(train, batch_size=BATCH_SIZE,
                                           shuffle=True, num_workers=2)
test_loader = torch.utils.data.DataLoader(test, batch_size=BATCH_SIZE,
                                          shuffle=True, num_workers=2)
```

```python
dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images.numpy()

img = np.squeeze(images[5])

fig = plt.figure(figsize = (6,6)) 

ax = fig.add_subplot(111)
ax.imshow(img[0])

labels = labels.numpy()

lbl = np.squeeze(labels[5])

fig = plt.figure(figsize = (6,6)) 

ax = fig.add_subplot(111)
ax.imshow(lbl)
```

```python
class Net(nn.Module):

  def __init__(self):
    super(Net, self).__init__()

    #Encoder

    #Conv layer 1 with a filter size of 3x3
    self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1)
    #Pooling layer 1
    self.pool1 = nn.MaxPool2d(2, 2)

    #Conv layer 2 with a filter size of 3x3
    self.conv2 = nn.Conv2d(32, 16, 3, padding=1)
    #Pooling layer 2
    self.pool2 = nn.MaxPool2d(2, 2)

    #Conv layer 3 with a filter size of 3x3
    self.conv3 = nn.Conv2d(16, 4, 3, padding=1)
    #Pooling layer 3
    self.pool3 = nn.MaxPool2d(2, 2)

    #Conv layer 4 with a filter size of 3x3
    self.conv4 = nn.Conv2d(4, 2, 3, padding=1)
    #Pooling layer 4
    self.pool4 = nn.MaxPool2d(2, 2)
    
    #Decoder
    self.t_conv1 = nn.ConvTranspose2d(in_channels=2, out_channels=4, kernel_size=2, stride=2)
    self.t_conv2 = nn.ConvTranspose2d(4, 16, 2, stride=2)
    self.t_conv3 = nn.ConvTranspose2d(16, 32, 2, stride=2)
    self.t_conv4 = nn.ConvTranspose2d(32, 1, 2, stride=2)


  def forward(self, x):
      #Activation
      x = F.relu(self.conv1(x))
      x = self.pool1(x)
      
      x = F.relu(self.conv2(x))
      x = self.pool2(x)

      x = F.relu(self.conv3(x))
      x = self.pool3(x)

      x = F.relu(self.conv4(x))
      x = self.pool4(x)

#decode
      x = F.relu(self.t_conv1(x))
      x = F.relu(self.t_conv2(x))
      x = F.relu(self.t_conv3(x))
      x = F.relu_(self.t_conv4(x))
            
      return x
```

```python
model = Net()
model
```

```python
cost_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr= LEANRING_RATE)
```

```python
for data in train_loader:
      images, labels = data
      output = model(images)
      labels = labels.to(torch.float32)
      print(output.dtype)
      print(images.dtype)
      print(labels.dtype)
      break;
```

```python
for epoch in range(NUM_EPOCHS):

  train_loss = 0.0

  for data in train_loader:
      images, labels = data

      output = model(images)
      labels = labels.to(torch.float32)
      cost = cost_function(output, labels)

      optimizer.zero_grad()
      cost.backward()
      optimizer.step()

      train_loss += cost.item()*images.size(0)
          
  train_loss = train_loss/len(train_loader)
  print('Epoch[{}/{}], loss: {:.6f}'
        .format(epoch + 1, NUM_EPOCHS, train_loss))
```

```python
dataiter = iter(test_loader)
images, labels = dataiter.next()
output = model(images)
images = images.numpy()

img = np.squeeze(images[5])

fig = plt.figure(figsize = (6,6)) 

ax = fig.add_subplot(111)
ax.imshow(img[0])

labels = labels.numpy()

lbl = np.squeeze(labels[5])

fig = plt.figure(figsize = (6,6)) 

ax = fig.add_subplot(111)
ax.imshow(lbl)


output = output.detach().numpy()
output = np.squeeze(output[5])

fig = plt.figure(figsize = (6,6)) 

ax = fig.add_subplot(111)
ax.imshow(output)
```
